```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# The abve needs to be converted into this form: ['Action','Adventure','Fantasy','Science Fiction']
# For this we need to create a helper function (convert) which will append the string under "name"
```

```python
# in the convert function obj must be integer so we have to call an another function which will first convert the 
# string of list in to list of string 
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List    
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# From cast column we require only first four lead cast name from the first four dictionary i.e dcount.
```

```python
def convertcast(obj):
    List = []
    dcount = 0
    for i in ast.literal_eval(obj):
        if dcount != 4:
         List.append(i['name'])
         dcount+=1
        else:
            break     
    return List 
```

```python
movies['cast'].apply(convertcast)
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
# In crew column we have to consider the dictionary which has the director's name.
```

```python
def search(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
         List.append(i['name'])
         break
    return List 
```

```python
movies['crew'].apply(search)
```

```python
movies['crew'] = movies['crew'].apply(search)
```

```python
# In the above tag there may be some similar words which needs to be removed.
# So we apply stemmer operation.
# eg: ['loved','loving','love'] ---> ['love','love','love']
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x = []
    
    for i in text.split():
        x.append(ps.stem(i))
        
    return " ".join(x)    
```

```python
movies_dataframe['tags'].apply(stem)
```

```python
movies_dataframe['tags'][0]
```

```python
movies_dataframe['tags'] = movies_dataframe['tags'].apply(lambda x:x.lower())
```

```python
# Now we have to find out the similarity between the different tags in the column.
# For this vectorization of the tags is required i.e textvectorization, by using "bag of words" technique.
# "Bag of words": Step1:- Combine all the tags and make a large text and from that text.
# Step2:- Find out frequency of most common words say 2000 thus matrix of 4806 x 2000 will form.
# Thus each movie will become a vector with 2000 dimentional space.
# Each tag will become a vector.Thus all the movie_id will be depicted by a vector.
# The vectors nearer(angular value less) to each other will have similarity.

# Before vectorization we have to remove the stop words from the tags eg:are,and,we,of,to,from,them etc.
# We will be using sklearn library for performing the above.We will use max feature
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=2000, stop_words='english')
```

```python
cv.fit_transform(movies_dataframe['tags']).toarray()
```

```python
cv.fit_transform(movies_dataframe['tags']).toarray().shape
```

```python
movies_vector = cv.fit_transform(movies_dataframe['tags']).toarray()
```

```python
# In the above 2000 words there may be some similar words which needs to be removed.
# So we apply steming operation.
# eg: ['loved','loving','love'] ---> ['love','love','love']
```

```python
# Now we have to calculate the angular distance (cosine distance) of refferance vector with the remaining vectors.
# With the help of sklearn function.
# Thus a matrix of 4806 x 4806 will form with diagonal value 1.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(movies_vector)
```

```python
cosine_similarity(movies_vector).shape
```

```python
movies_similarity = cosine_similarity(movies_vector)
```

```python
sorted(list(enumerate(movies_similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
movies_dataframe[movies_dataframe['title'] == 'Batman Begins'].index[0]
```

```python
def recommended_movies(movie_input):
    movies_index = movies_dataframe[movies_dataframe['title'] == movie_input].index[0]
    distances = movies_similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_dataframe.iloc[i[0]].title)
```

```python
recommended_movies('Batman Begins')
```
